```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset("titanic")
```

```python

# 1. Visualize total survivors vs. non-survivors using a bar chart
sns.countplot(x="survived", data=titanic,hue="survived", palette="viridis",legend=False)
plt.title("Total Survivors vs Non-Survivors")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

**Observation:**  
More people did **not survive (0)** compared to those who **survived (1)**.  
The dataset is clearly imbalanced toward non-survivors

```python
# 2. Visualize survival rate differences between male and female passengers
sns.barplot(x="sex", y="survived", data=titanic,hue="sex", palette="pastel",legend=False)
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

**Observation:**  
Female passengers had a much **higher survival rate** compared to males.  
This suggests the **"women and children first"** policy influenced survival strongly.


```python
# 3. Use Heatmaps to visualize survival rate by passenger class

heatmap_data = titanic.pivot_table("survived", index="sex", columns="class", observed=False)
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Survival Rate by Gender and Class")
plt.show()
```

**Observation:**  
- **Highest survival**: Females in **1st class**  
- **Lowest survival**: Males in **3rd class**  
Both **gender** and **passenger class** strongly influenced survival chances.


```python
# 4. Compare survival counts across 1st, 2nd, and 3rd classes
sns.countplot(x="class", hue="survived", data=titanic, palette="Set2")
plt.title("Survival Counts by Passenger Class")
plt.show()
```

**Observation:**  
- More passengers in **3rd class died** compared to 1st and 2nd class.  
- Passengers in **higher classes had better chances of survival.**


```python
# 5. Use histograms, KDE, or violin plots to compare ages of survivors vs. non-survivors
sns.violinplot(x="survived", y="age",hue="survived", data=titanic, palette="muted",legend= False)
plt.title("Age Distribution of Survivors vs Non-Survivors")
plt.show()
```

**Observation:**  
- Younger passengers, especially **children**, had better chances of survival.  
- Among **older age groups**, survival rates decreased significantly.


```python
# 6. Use Box plots to visualize age distribution by survival status and detect outliers
sns.boxplot(x="survived", y="age", data=titanic,hue ="survived", palette="Set3",legend =False)
plt.title("Boxplot of Age by Survival Status")
plt.show()
```

**Observation:**  
- Survivors were generally **younger** than non-survivors.  
- A few **outliers (very old passengers)** also survived, though rare.



# Final Insights

- **Gender**: Females had higher survival rates than males.  
- **Class**: Higher-class passengers were more likely to survive.  
- **Age**: Younger passengers (especially children) had better survival chances.  
- The dataset confirms that **social status (class + gender) and age** were critical factors in survival on the Titanic.
