```python
import matplotlib.pyplot as plt
from PIL import Image   # ✅ you need this for Image.open

# List of image file names (keep them in the same folder as your notebook)
images = ["IMG1.jpg", "IMG2.jpg", "IMG3.jpg", "IMG4.jpg","IMG5.jpg", "IMG6.jpg"]

# Open images
imgs = [Image.open(img) for img in images]

# Resize them to a common size for better collage
imgs = [img.resize((300, 300)) for img in imgs]

# Create a blank white canvas (adjust size as per number of images)
collage_width = 900   # 3 images wide (3 × 300)
collage_height = 600  # 2 images tall (2 × 300)
collage = Image.new("RGB", (collage_width, collage_height), "white")

# Paste images into the canvas
x_offset = 0
y_offset = 0
count = 0
for img in imgs:
    collage.paste(img, (x_offset, y_offset))
    x_offset += 300   # move right
    count += 1
    if count % 3 == 0:   # after 3 images, move to next row
        x_offset = 0
        y_offset += 300

# ✅ Show the final collage
plt.figure(figsize=(10, 7))
plt.imshow(collage)
plt.axis("off")
plt.show()
```
