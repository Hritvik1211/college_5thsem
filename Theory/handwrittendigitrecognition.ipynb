```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load MNIST dataset using Keras
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Display the first 10 images
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
axes = axes.flatten()

for i in range(10):
    axes[i].imshow(X_train[i], cmap='gray')
    axes[i].set_title(f'Label: {y_train[i]}')
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
# PCA
X_train_flat = X_train.reshape(X_train.shape[0], -1)

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_flat)

# Perform PCA to 2D
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)

# Scatter plot of the first two principal components colored by label
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_train_pca[:, 0],
    X_train_pca[:, 1],
    c=y_train,
    cmap='tab10',
    s=3,
    alpha=0.6
)
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
plt.title('PCA of MNIST Digits (2D)')
plt.grid(True, alpha=0.3)
plt.colorbar(scatter, ticks=range(10), label='Digit Label')
plt.tight_layout()
plt.show()

# Visualize the explained variance ratio
plt.figure(figsize=(10, 6))
pca_full = PCA().fit(X_train_scaled)
plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Components')
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import seaborn as sns
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten

# Prepare the data
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_flat)
X_test_scaled = scaler.transform(X_test_flat)

# For faster processing, use a subset of the data (optional)
# Remove this block if you want to use the full dataset
subset_size = 10000
X_train_subset = X_train_scaled[:subset_size]
y_train_subset = y_train[:subset_size]
X_test_subset = X_test_scaled[:1000]
y_test_subset = y_test[:1000]

# Function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print(classification_report(y_true, y_pred))

# 1. Random Forest
print("Training Random Forest...")
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_model.fit(X_train_subset, y_train_subset)
rf_preds = rf_model.predict(X_test_subset)
plot_confusion_matrix(y_test_subset, rf_preds, "Random Forest Confusion Matrix")

# 2. SVM
print("Training SVM...")
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(X_train_subset, y_train_subset)
svm_preds = svm_model.predict(X_test_subset)
plot_confusion_matrix(y_test_subset, svm_preds, "SVM Confusion Matrix")

# 3. Neural Network using sklearn
print("Training Neural Network (sklearn)...")
mlp_model = MLPClassifier(hidden_layer_sizes=(128,64), max_iter=20, alpha=0.0001,
                          solver='adam', random_state=42, early_stopping=True)
mlp_model.fit(X_train_subset, y_train_subset)
mlp_preds = mlp_model.predict(X_test_subset)
plot_confusion_matrix(y_test_subset, mlp_preds, "Neural Network (sklearn) Confusion Matrix")

# 4. Neural Network using Keras
print("Training Neural Network (Keras)...")
# Convert the labels to one-hot encoding for Keras
y_train_cat = to_categorical(y_train_subset, 10)
y_test_cat = to_categorical(y_test_subset, 10)

# Create a simple neural network
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_subset.shape[1],)),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_subset, y_train_cat, epochs=10, batch_size=128, validation_split=0.1, verbose=1)

# Get predictions and convert back from one-hot encoding
keras_preds = model.predict(X_test_subset)
keras_preds = np.argmax(keras_preds, axis=1)

plot_confusion_matrix(y_test_subset, keras_preds, "Neural Network (Keras) Confusion Matrix")
```

```python
# Model performance comparison
import pandas as pd
from sklearn.metrics import f1_score

# Ensure required variables exist: y_test_subset, rf_preds, svm_preds, mlp_preds, keras_preds
models_predictions = {
    'Random Forest': rf_preds,
    'SVM (RBF)': svm_preds,
    'MLP (sklearn)': mlp_preds,
    'Neural Net (Keras)': keras_preds,
}

comparison_rows = []
for model_name, y_pred in models_predictions.items():
    acc = accuracy_score(y_test_subset, y_pred)
    f1_macro = f1_score(y_test_subset, y_pred, average='macro')
    f1_weighted = f1_score(y_test_subset, y_pred, average='weighted')
    comparison_rows.append({
        'Model': model_name,
        'Accuracy': acc,
        'F1_macro': f1_macro,
        'F1_weighted': f1_weighted,
    })

comparison_df = pd.DataFrame(comparison_rows).sort_values('Accuracy', ascending=False).reset_index(drop=True)
comparison_df
```

```python
# Bar chart of accuracies
plt.figure(figsize=(8, 5))
plt.bar(comparison_df['Model'], comparison_df['Accuracy'], color=['#4c72b0','#55a868','#c44e52','#8172b2'])
plt.ylim(0.85, 1.0)
plt.ylabel('Accuracy')
plt.title('Model Accuracy on MNIST Subset')
plt.xticks(rotation=15)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
